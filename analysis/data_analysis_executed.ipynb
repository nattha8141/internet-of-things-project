## 2. Data Cleaning

### 2.1 Identifying Anomalies
Some data points show extreme simultaneous drops across all three photoresistor channels. These are caused by physical disturbances during watering (e.g., unintentionally shifting the sensor or touching the breadboard). We identify these as rows where **all three** sensor values change drastically at once.

````python
# Compute absolute differences between consecutive readings
diff_green = df['Green'].diff().abs()
diff_blue = df['Blue'].diff().abs()
diff_control = df['Control'].diff().abs()

# Anomaly: all three channels jump by more than 200 simultaneously
THRESHOLD = 200
anomaly_mask = (diff_green > THRESHOLD) & (diff_blue > THRESHOLD) & (diff_control > THRESHOLD)

# Also check for the recovery point (next row after anomaly)
anomaly_indices = df.index[anomaly_mask].tolist()
# Include the anomalous rows themselves
rows_to_remove = set()
for idx in anomaly_indices:
    rows_to_remove.add(idx)
    # Check if previous row is also anomalous (the disturbance point)
    if idx > 0:
        prev_vals = df.loc[idx-1, ['Green', 'Blue', 'Control']].values
        curr_vals = df.loc[idx, ['Green', 'Blue', 'Control']].values
        # If previous row values are way off from the trend, remove it too
        if idx - 1 not in anomaly_indices:
            rows_to_remove.add(idx - 1)

print(f"Anomalous rows detected: {len(rows_to_remove)}")
print("\nAnomalous data points:")
for idx in sorted(rows_to_remove):
    print(f"  Row {idx}: {df.loc[idx, 'Timestamp']} | G={df.loc[idx, 'Green']} B={df.loc[idx, 'Blue']} C={df.loc[idx, 'Control']}")
````

````python
# Visualise anomalies on raw data
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(df['Timestamp'], df['Green'], 'g-', alpha=0.7, label='Green', linewidth=0.8)
ax.plot(df['Timestamp'], df['Blue'], 'b-', alpha=0.7, label='Blue', linewidth=0.8)
ax.plot(df['Timestamp'], df['Control'], 'k-', alpha=0.7, label='Control', linewidth=0.8)

# Mark anomalies
for idx in sorted(rows_to_remove):
    ax.axvline(df.loc[idx, 'Timestamp'], color='red', alpha=0.5, linestyle='--', linewidth=0.8)

ax.set_xlabel('Time')
ax.set_ylabel('Photoresistor Reading (ADC)')
ax.set_title('Raw Sensor Data with Anomalies Highlighted (Red Lines)')
ax.legend()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.tight_layout()
plt.savefig(IMAGE_DIR + 'raw_data_with_anomalies.png', dpi=150, bbox_inches='tight')
plt.show()
````

````python
# Remove anomalies and interpolate
df_clean = df.copy()
for idx in rows_to_remove:
    df_clean.loc[idx, ['Green', 'Blue', 'Control']] = np.nan

# Interpolate missing values
df_clean[['Green', 'Blue', 'Control']] = df_clean[['Green', 'Blue', 'Control']].interpolate(method='linear')

print(f"Cleaned dataset: {df_clean.shape[0]} rows ({len(rows_to_remove)} anomalous points interpolated)")
print(f"Remaining NaNs: {df_clean[['Green', 'Blue', 'Control']].isna().sum().sum()}")
````

### 3.4 Plant Movement / ADC Drop Events

All three groups exhibit periodic drops in ADC readings. **None of these drops indicate the plant shrinking** — they are all caused by different forms of plant movement:

- **Control (dark)**: Sprouts bend toward blue light leaking from the adjacent chamber (**phototropism**). This moves them out of the sensor path (reading drops). New growth reaches back into the path (reading rises), creating oscillation.

- **Blue & Green (illuminated)**: Sprouts grow toward the LED (the light source). Initially, upward growth blocks the sensor (ADC rises). But once stems **outgrow the confined chamber**, they become top-heavy and disarray — flopping, tangling, or bending against walls. Leaves that were blocking the sensor shift aside, causing sudden ADC drops. New growth fills the gap and readings recover.

This explains why the blue channel shows intermittent drops from 4095 even after saturation — the plant kept growing but periodically rearranged itself within the limited space.

````python
# Detrend the control signal to isolate oscillation
# Use a large rolling mean as the trend
trend_window = 48  # 12-hour trend
df_clean['Control_trend'] = df_clean['Control'].rolling(trend_window, center=True).mean()
df_clean['Control_detrended'] = df_clean['Control'] - df_clean['Control_trend']

# Also detrend green and blue for comparison
df_clean['Green_trend'] = df_clean['Green'].rolling(trend_window, center=True).mean()
df_clean['Green_detrended'] = df_clean['Green'] - df_clean['Green_trend']
df_clean['Blue_trend'] = df_clean['Blue'].rolling(trend_window, center=True).mean()
df_clean['Blue_detrended'] = df_clean['Blue'] - df_clean['Blue_trend']

fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

for ax, col, color, name in zip(axes, 
    ['Green_detrended', 'Blue_detrended', 'Control_detrended'],
    ['#2ca02c', '#1f77b4', '#333333'],
    ['Green Light', 'Blue Light', 'Control (Dark)']):
    ax.plot(df_clean['Timestamp'], df_clean[col], color=color, linewidth=0.8)
    ax.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
    ax.set_ylabel('Detrended ADC')
    ax.set_title(f'{name} — Detrended Signal (Oscillation / Plant Movement)')
    ax.grid(True, alpha=0.3)

axes[-1].set_xlabel('Time')
axes[-1].xaxis.set_major_formatter(mdates.DateFormatter('%b %d\n%H:%M'))
plt.tight_layout()
plt.savefig(IMAGE_DIR + 'plant_movement_oscillation.png', dpi=150, bbox_inches='tight')
plt.show()
````

### 3.5 Geometric Height Estimation Model

Since no direct height sensor was available, we can **approximate plant height from the photoresistor ADC readings** using the physical geometry of the setup.

#### Setup Geometry (Side View)
```
    ┌─────────────────────────────┐
    │                         [LED]│  ← H_led (~11 cm from floor)
    │                        ╱    │
    │                      ╱      │  Light cone
    │        ╱╲ ╱╲       ╱        │
    │       ╱  ╳  ╲    ╱          │
    │      │  │ │  │  ╱           │
    │ [LDR]│  │ │  │╱             │  ← H_sensor (~4 cm)
    │      └──┴─┴──┘              │
    │       Plant Container       │  ← H_base (~4 cm, container rim)
    └─────────────────────────────┘
         ←─ D_h (~10 cm) ────────→
```

**Model**: As the plant grows from the container rim, it progressively blocks the light cone between the LED and the photosensor. We define:

- `obstruction_ratio = (ADC - ADC_baseline) / (ADC_max - ADC_baseline)` → ranges from 0 (no growth) to 1 (fully blocked)
- Using the chamber geometry (LED height, sensor height, horizontal distance), we calculate the **critical height** at which the plant fully blocks the direct line of sight
- The estimated height is then: `h_est = obstruction_ratio × H_full_block`

**Key assumption**: The relationship between obstruction ratio and height is approximately linear. This is a first-order approximation — in reality, the relationship depends on plant density, leaf spread, and how centrally the plant sits in the light cone.

**Calibration**: For the blue channel, we know ADC reached 4095 (full saturation), which serves as a calibration anchor point.

````python
# ============================================================
# GEOMETRIC HEIGHT ESTIMATION MODEL
# ============================================================
# Adjustable physical parameters (estimated from photos of the Dune London shoebox setup)
# Users can refine these measurements for better accuracy.

H_led = 11.0      # LED height above box floor (cm) — taped near top corner
H_sensor = 4.0     # Photosensor height above box floor (cm) — at container rim
D_horizontal = 10.0  # Horizontal distance between LED and sensor (cm)
H_base = 4.0       # Container rim height (cm) — plant growth starts here
X_plant = 5.0      # Horizontal position of plant center from sensor (cm)

# --- Calculate the light path geometry ---
# The direct line of sight from LED to sensor at the plant's horizontal position:
#   h_line(x) = H_sensor + (H_led - H_sensor) * (x / D_horizontal)
# At plant position:
h_line_at_plant = H_sensor + (H_led - H_sensor) * (X_plant / D_horizontal)

# The plant needs to grow above H_base. The height above the container rim
# that fully blocks the direct line of sight:
H_full_block_direct = h_line_at_plant - H_base

# But the LED has a spread angle. For full ADC saturation (4095), the plant
# must block not just the direct line but enough of the cone. We add a
# margin factor (~1.5x) to account for this:
CONE_FACTOR = 1.5
H_full_block = H_full_block_direct * CONE_FACTOR

print("=== Geometric Height Model Parameters ===")
print(f"  LED height:           {H_led} cm")
print(f"  Sensor height:        {H_sensor} cm")
print(f"  Horizontal distance:  {D_horizontal} cm")
print(f"  Container rim:        {H_base} cm")
print(f"  Plant position:       {X_plant} cm from sensor")
print(f"  Line-of-sight height at plant: {h_line_at_plant:.1f} cm")
print(f"  Direct block height (above rim): {H_full_block_direct:.1f} cm")
print(f"  Full block height with cone factor ({CONE_FACTOR}x): {H_full_block:.1f} cm")
print(f"\n  → When ADC reaches 4095, estimated plant height ≈ {H_full_block:.1f} cm above rim")

# --- Compute estimated heights for each channel ---
baselines = {
    'Green': df_clean['Green'].iloc[:10].mean(),
    'Blue': df_clean['Blue'].iloc[:10].mean(),
    'Control': df_clean['Control'].iloc[:10].mean()
}

ADC_MAX = 4095

for ch in ['Green', 'Blue', 'Control']:
    baseline = baselines[ch]
    # Obstruction ratio: 0 = no growth, 1 = fully blocked
    ratio = (df_clean[ch] - baseline) / (ADC_MAX - baseline)
    # Clamp to [0, 1] — values > 1 are sensor saturation, < 0 is movement
    ratio = ratio.clip(lower=0, upper=1)
    df_clean[f'{ch}_height_cm'] = ratio * H_full_block

print(f"\n=== Estimated Heights at End of Experiment ===")
for ch, color_name in [('Green', 'Green'), ('Blue', 'Blue'), ('Control', 'Control')]:
    h = df_clean[f'{ch}_height_cm'].iloc[-1]
    h_max = df_clean[f'{ch}_height_cm'].max()
    print(f"  {color_name:8s}: Current ≈ {h:.1f} cm | Max observed ≈ {h_max:.1f} cm")
````

````python
# --- Plot: Estimated Height Growth Curves ---
fig, axes = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# Top: Estimated height in cm
ax1 = axes[0]
ax1.plot(df_clean['Timestamp'], df_clean['Green_height_cm'], color='#2ca02c', label='Green Light', linewidth=1.2)
ax1.plot(df_clean['Timestamp'], df_clean['Blue_height_cm'], color='#1f77b4', label='Blue Light', linewidth=1.2)
ax1.plot(df_clean['Timestamp'], df_clean['Control_height_cm'], color='#333333', label='Control (Dark)', linewidth=1.2)
ax1.axhline(y=H_full_block, color='red', linestyle=':', alpha=0.5, label=f'Sensor limit ({H_full_block:.1f} cm)')
ax1.set_ylabel('Estimated Height (cm above rim)')
ax1.set_title(f'Estimated Plant Height from Geometric Model (H_full_block = {H_full_block:.1f} cm)')
ax1.legend(loc='upper left')
ax1.grid(True, alpha=0.3)
ax1.set_ylim(bottom=-0.5)

# Bottom: Growth rate in cm/hour
for ch in ['Green', 'Blue', 'Control']:
    df_clean[f'{ch}_height_rate'] = df_clean[f'{ch}_height_cm'].diff(4) / 1.0  # 4 samples = 1 hour
    df_clean[f'{ch}_height_rate_smooth'] = df_clean[f'{ch}_height_rate'].rolling(8, center=True).mean()

ax2 = axes[1]
ax2.plot(df_clean['Timestamp'], df_clean['Green_height_rate_smooth'], color='#2ca02c', label='Green', linewidth=1.2)
ax2.plot(df_clean['Timestamp'], df_clean['Blue_height_rate_smooth'], color='#1f77b4', label='Blue', linewidth=1.2)
ax2.plot(df_clean['Timestamp'], df_clean['Control_height_rate_smooth'], color='#333333', label='Control', linewidth=1.2)
ax2.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
ax2.set_ylabel('Growth Rate (cm/hour)')
ax2.set_xlabel('Time')
ax2.set_title('Estimated Growth Rate (cm/hour, 2-hour smoothed)')
ax2.legend(loc='upper left')
ax2.grid(True, alpha=0.3)

for ax in axes:
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d\n%H:%M'))

plt.tight_layout()
plt.savefig(IMAGE_DIR + 'estimated_height_growth.png', dpi=150, bbox_inches='tight')
plt.show()

# Print growth rate summary
print("=== Estimated Growth Rates (cm/hour) ===")
for ch in ['Green', 'Blue', 'Control']:
    rate = df_clean[f'{ch}_height_rate'].dropna()
    positive_rate = rate[rate > 0]  # Only positive growth
    print(f"  {ch:8s}: Mean = {rate.mean():.3f} cm/h | Peak = {rate.max():.3f} cm/h | Mean (positive only) = {positive_rate.mean():.3f} cm/h")
````

````python
# --- Visualise the geometric model: ADC → Height calibration curve ---
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: ADC to Height mapping
adc_range = np.linspace(3400, 4095, 200)
for ch, color, bl in [('Green', '#2ca02c', baselines['Green']), 
                       ('Blue', '#1f77b4', baselines['Blue']),
                       ('Control', '#333333', baselines['Control'])]:
    ratio = np.clip((adc_range - bl) / (ADC_MAX - bl), 0, 1)
    height = ratio * H_full_block
    axes[0].plot(adc_range, height, color=color, linewidth=2, label=f'{ch} (baseline={bl:.0f})')

axes[0].set_xlabel('Photoresistor ADC Reading')
axes[0].set_ylabel('Estimated Height (cm)')
axes[0].set_title('ADC → Estimated Height Calibration Curve')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].axvline(x=4095, color='red', linestyle=':', alpha=0.5)

# Right: Side-view geometry diagram
ax2 = axes[1]
ax2.set_xlim(-1, D_horizontal + 1)
ax2.set_ylim(-1, H_led + 2)
ax2.set_aspect('equal')

# Box outline
box = plt.Rectangle((0, 0), D_horizontal, H_led + 1, fill=False, edgecolor='black', linewidth=2)
ax2.add_patch(box)

# LED
ax2.plot(D_horizontal - 0.5, H_led, 'r^', markersize=12, label='LED (RGB)')
ax2.annotate('LED', (D_horizontal - 0.5, H_led), textcoords="offset points", xytext=(10, 5), fontsize=9)

# Sensor
ax2.plot(0.5, H_sensor, 'bs', markersize=10, label='Photosensor')
ax2.annotate('Sensor', (0.5, H_sensor), textcoords="offset points", xytext=(10, -15), fontsize=9)

# Light path (direct line)
ax2.plot([D_horizontal - 0.5, 0.5], [H_led, H_sensor], 'y-', linewidth=1.5, alpha=0.6, label='Light path')

# Light cone (approximate)
ax2.fill([D_horizontal - 0.5, 0.5, 0.5, D_horizontal - 0.5], 
         [H_led, H_sensor + 1.5, H_sensor - 1, H_led], 
         alpha=0.08, color='yellow')

# Plant container
container = plt.Rectangle((X_plant - 1.5, 0), 3, H_base, fill=True, facecolor='#8B4513', alpha=0.3, edgecolor='brown')
ax2.add_patch(container)
ax2.annotate('Container', (X_plant, H_base / 2), ha='center', fontsize=8)

# Plant (vertical lines from container up to estimated full block height)
for dx in [-0.5, 0, 0.5]:
    plant_x = X_plant + dx
    ax2.plot([plant_x, plant_x], [H_base, H_base + H_full_block], 'g-', linewidth=2, alpha=0.5)
    # Leaf at top
    ax2.plot(plant_x, H_base + H_full_block, 'g^', markersize=6)

# Height markers
ax2.annotate('', xy=(X_plant + 2, H_base), xytext=(X_plant + 2, H_base + H_full_block),
            arrowprops=dict(arrowstyle='<->', color='red', lw=1.5))
ax2.annotate(f'H_block\n≈{H_full_block:.1f}cm', (X_plant + 2.2, H_base + H_full_block/2), 
            fontsize=9, color='red', va='center')

ax2.set_xlabel('Horizontal (cm)')
ax2.set_ylabel('Height (cm)')
ax2.set_title('Chamber Side-View Geometry')
ax2.legend(loc='upper left', fontsize=8)

plt.tight_layout()
plt.savefig(IMAGE_DIR + 'geometric_model.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\nNote: The control group shows negative estimated heights due to phototropic")
print(f"bending — the sprouts move AWAY from the sensor path, not shrinking.")
print(f"The height model is most accurate for the blue and green groups where")
print(f"the sprouts grow predominantly upward through the light cone.")
````

## 4. Statistical Analysis

### 4.1 Growth Rate Comparison (t-tests)

````python
# Compare growth rates between groups (using non-null rates)
rates = df_clean[['Green_rate', 'Blue_rate', 'Control_rate']].dropna()

# Welch's t-test: Green vs Blue
t_gb, p_gb = stats.ttest_ind(rates['Green_rate'], rates['Blue_rate'], equal_var=False)
t_gc, p_gc = stats.ttest_ind(rates['Green_rate'], rates['Control_rate'], equal_var=False)
t_bc, p_bc = stats.ttest_ind(rates['Blue_rate'], rates['Control_rate'], equal_var=False)

print("=== Welch's t-test on Hourly Growth Rates ===")
print(f"Green vs Blue:    t={t_gb:.3f}, p={p_gb:.4f} {'***' if p_gb < 0.001 else '**' if p_gb < 0.01 else '*' if p_gb < 0.05 else 'ns'}")
print(f"Green vs Control: t={t_gc:.3f}, p={p_gc:.4f} {'***' if p_gc < 0.001 else '**' if p_gc < 0.01 else '*' if p_gc < 0.05 else 'ns'}")
print(f"Blue vs Control:  t={t_bc:.3f}, p={p_bc:.4f} {'***' if p_bc < 0.001 else '**' if p_bc < 0.01 else '*' if p_bc < 0.05 else 'ns'}")

print(f"\nMean growth rates (ADC/15min):")
print(f"  Green:   {rates['Green_rate'].mean():.3f} ± {rates['Green_rate'].std():.3f}")
print(f"  Blue:    {rates['Blue_rate'].mean():.3f} ± {rates['Blue_rate'].std():.3f}")
print(f"  Control: {rates['Control_rate'].mean():.3f} ± {rates['Control_rate'].std():.3f}")
````

## 5. Key Findings Summary

1. **Blue light accelerates growth the most**: The blue-light group saturated the ADC sensor (4095) within ~3 days, corresponding to ~5.2 cm estimated height. Manual observation confirmed green, chlorophyll-rich leaves and the tallest sprouts.

2. **Green light has moderate but sensor-underrepresented growth**: The green group showed ~49 ADC change (~0.4 cm estimated), but this underestimates actual growth because ambient green light leaks through to the sensor. Visual observation confirmed decent growth.

3. **ADC drops ≠ plant shrinkage**: All three groups show periodic ADC drops, but none indicate reduced growth:
   - **Control**: Phototropic bending toward blue light leak moves sprouts out of the sensor path.
   - **Blue/Green**: Once stems outgrow the chamber, they become top-heavy and disarray — leaves shift aside, momentarily unblocking the sensor. New growth then fills the gap.

4. **Geometric height model bridges ADC to physical units**: Using chamber geometry (LED/sensor positions, distances), we converted raw ADC readings to estimated plant height in cm — enabling physically meaningful growth rate calculations (cm/hour).

5. **Inter-chamber correlation has no causal meaning**: The three chambers are isolated. Any observed ADC correlation reflects shared environmental factors (temperature, time of day), not direct interaction.

6. **Environmental conditions were stable**: Temperature ~24.6–27.1°C, humidity ~54–62%. Weak correlation with growth (|r| < 0.15) confirms light treatment is the primary driver.

7. **Leaf coloration confirms biological effect**: Blue → green leaves (chlorophyll), Green → slight green, Control → yellow (etiolated). A future colour sensor (TCS34725) could quantify this.
